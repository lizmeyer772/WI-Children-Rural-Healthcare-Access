Dataset Manipulation and Cleaning for NSCH.csv

```python
import pandas as pd
```

```python
NSCH = pd.read_csv("NSCH.csv")
```

Has an MEDB (1 = Has MEDB, 2 = Does not have MEDB)

```python
NSCH["MEDB"] = NSCH["MEDB10ScrQ5_22"].replace(95, pd.NA)
NSCH = NSCH.dropna(subset=["MEDB"])
```

Urban/Rural Area (1 = Urban, 2 = Rural)

```python
NSCH["Metro"] = NSCH["METRO_YN"].replace(96, pd.NA)
NSCH = NSCH.dropna(subset=["Metro"])
```

Subset to only children with an MEDB

```python
NSCHUS = NSCH[NSCH["MEDB"] == 1].copy()
```

Health insurance status (1 = Insured, 2 = Uninsured)

```python
NSCHUS["Ins"] = NSCHUS["CurrIns_22"].replace(99, pd.NA)
NSCHUS = NSCHUS.dropna(subset=["Ins"])
```

Type of health insurance
(1 = Public, 2 = Private, 3 = Both, 4 = Uninsured)

```python
NSCHUS["InsTypes"] = NSCHUS["instype_22"].replace(99, pd.NA)
NSCHUS = NSCHUS.dropna(subset=["InsTypes"])
```

Caregiver (CG) employment status (1 = at least 1 CG employed full-time,
2 = at least 1 CG employed part-time, 3 = CG(s) unemployed)

```python
NSCHUS["CGEmploy"] = NSCHUS["EmploymentSt_22"].replace(99, pd.NA)
NSCHUS = NSCHUS.dropna(subset=["CGEmploy"])
```

Family structure (1 = 2 married parents, 2 = 2 unmarried parents, 3 = single parent,
4 = Grandparent, 5 = Other)

```python
NSCHUS["Family"] = NSCHUS["famstruct5_22"].replace(99, pd.NA)
NSCHUS = NSCHUS.dropna(subset=["Family"])
```

Federal poverty level (FPL) (1 = 0-99% FPL, 2 = 100-199% FPL, 3 = 200-399% FPL,
4 = 400% FPL or greater)

```python
NSCHUS["FPL"] = NSCHUS["povlev4_22"]
```

Sex of child (1 = Male, 2 = Female)

```python
NSCHUS["Sex"] = NSCHUS["sex_22"]
```

Race/ethnicity of child (1 = Hispanic, 2 = White, 3 = Black, 4 = Asian,
5 = American Indian/Alaska, 6 = Native Hawaiian and Other PI, 7 = Multi-Race)

```python
NSCHUS["Race"] = NSCHUS["race7_22"]
```

Age group of child (1 = 0-5, 2 = 6-11, 3 = 12-17)

```python
NSCHUS["Age"] = NSCHUS["age3_22"]
```

Adverse Childhood Experiences (ACEs) (1 = None, 2 = 1 ACE, 3 = 2 or more ACEs)

```python
NSCHUS["Aces"] = NSCHUS["ACE2more11_22"].replace(99, pd.NA)
NSCHUS = NSCHUS.dropna(subset=["Aces"])
```

Received Mental Health Care in the past 12 months (0 = No, 1 = Yes)

```python
NSCHUS["RecMHC"] = (
    NSCHUS["MentHCare_22"]
    .replace(99, pd.NA)
    .replace(3, 2)
)
NSCHUS = NSCHUS.dropna(subset=["RecMHC"])
```

Recode outcome variable (0 = No, 1 = Yes)

```python
NSCHUS["RecMHC"] = NSCHUS["MentHCare_22"].apply(
    lambda x: 0 if x == 2 else 1
)
```

Final datasets

```python
cols = ["Ins", "InsTypes", "CGEmploy", "Family", "FPL", "Sex",
        "Race", "Age", "Aces", "Metro", "MEDB", "RecMHC"]
```

For all children with a MEDB problem

```python
FinalAll = NSCHUS.loc[NSCHUS["MEDB"] == 1, cols].copy()
for c in FinalAll.columns:
    if FinalAll[c].dtype != "object":
        FinalAll[c] = FinalAll[c].astype("category")
```

For all children with a MEDB problem AND who are rural

```python
FinalRural = NSCHUS.loc[(NSCHUS["Metro"] == 2) & (NSCHUS["MEDB"] == 1), cols].copy()
for c in FinalRural.columns:
    if FinalRural[c].dtype != "object":
        FinalRural[c] = FinalRural[c].astype("category")
```

```python
FinalAll.to_csv("FinalAll.csv", index=False)
FinalRural.to_csv("FinalRural.csv", index=False)
```
