Probability

```python
import pandas as pd
```

```python
FinalRural = pd.read_csv("FinalRural.csv")
```

```python
# N
Total = len(FinalRural)
```

Counts

```python
#Sum of 2+ Aces
TwoAces = (FinalRural["Aces"] == 2).sum()

#Sum of Below Poverty Line
BelowFPL = (FinalRural["FPL"] == 1).sum()

#Sum of 2+ Aces AND Below Poverty Line
TwoAcesBelowFPL = ((FinalRural["Aces"] == 2) & (FinalRural["FPL"] == 1)).sum()
```

Proportions

```python
#Proportion of 2+ Aces
PropAces = TwoAces / Total

#Proportion of Below Poverty Line
PropFPL = BelowFPL / Total

#Proportion of 2+ Aces AND Below Poverty Line
PropAcesFPL = TwoAcesBelowFPL / Total
```

Conditional Probability: P(2+ Aces | Below Poverty)

```python
#Probability of 2+ Aces GIVEN Below Poverty Line
CondProb = PropAcesFPL / PropFPL
```

```python
print("P(2+ Aces) =", PropAces)
print("P(Below Poverty) =", PropFPL)
print("P(2+ Aces AND Below Poverty) =", PropAcesFPL)
print("P(2+ Aces | Below Poverty) =", CondProb)
```
