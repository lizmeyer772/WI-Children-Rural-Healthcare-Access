Statistical Inference

```python
import pandas as pd
```

```python
CHR = pd.read_excel('CountyHealthNew.xlsx')
```

Subsets to counties with RUCC codes 4-9

```python
CHRRur = CHR[CHR["RUCC"].isin([4, 5, 6, 7, 8, 9])]
```

Ranks counties individually and then creates a combined rank variable

```python
Ranks = CHRRur.copy()

Ranks["MHPRank"] = Ranks["# Mental Health Providers"].rank(method="min")
Ranks["ChildPovRank"] = Ranks["% Children in Poverty"].rank(method="min")
Ranks["SPRank"] = Ranks["% Children in Single-Parent Households"].rank(method="min")

Ranks["CombinedRank"] = Ranks["MHPRank"] + Ranks["ChildPovRank"] + Ranks["SPRank"]

HighNeedCounties = (
    Ranks.sort_values("CombinedRank")
      .loc[:, [
          "County",
          "# Mental Health Providers",
          "% Children in Poverty",
          "% Children in Single-Parent Households",
          "% Less than 18 Years of Age",
          "CombinedRank",
          "RUCC",
      ]]
      .head(20)
)

HighNeedCounties.insert(
    0,
    "FinalRank",
    HighNeedCounties.reset_index(drop=True).index + 1
)

```
